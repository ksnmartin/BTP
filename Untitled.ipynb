```python
import numpy as np
from lorenz import lorenz
import matplotlib.pyplot as plt
from scipy.integrate import solve_ivp
from autoencoder import sindy
from pysindy import SINDy
```

```python
simulator = lorenz(iteration=10000, dt=0.01)
xp, yp, zp = 1,1.,1.05
```

```python
xs, ys, zs = simulator.simulate(xp, yp, zp)
fig = plt.figure()
ax = fig.gca(projection='3d')
ax.plot(xs, ys, zs, lw=0.5)
ax.set_xlabel("X")
ax.set_ylabel("Y")
ax.set_zlabel("Z")
ax.set_title("Lorenz Attractor")
plt.show()
```

```python
x= np.stack((xs, ys, zs),axis=-1)
model = SINDy()
t=np.arange(0,len(x),0.001)
model.fit(x,t=t[:len(x)])
model.print()
model.score(x,t=t[:len(x)]) # score is the same as the loss
```

```python
tv= np.arange(0,1,0.001)
xt = model.simulate(
    x[0],
    t=tv
)
```

```python
xfig,xplot = plt.subplots(1 ,1, figsize=(10,10))

xreal = [x[i][0] for i in range(600,1000)]
xpred = [xt[i][0] for i in range(600,1000)]
xplot.plot(xpred,np.arange(0.6,1,0.001),label='pred')
xplot.plot(xreal,np.arange(0.6,1,0.001),label='real')
plt.legend(loc="upper left")
plt.show()
```
